**Objective: Predict the percentage of a student based on the number of study hours using Supervised ML(Linear regression).**

```python
# @title Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# @title Load the dataset
data_url = "http://bit.ly/w-data"
d = pd.read_csv(data_url)
d.head(25)
```

```python
# @title Spliting the data
X = d.iloc[:, :-1].values #inputs
Y = d.iloc[:, 1].values #labels
```

```python
# @title Showing the data
plt.scatter(X,Y, c='purple', label='Data Points')
plt.title('Hours Studied vs Percentage Score')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.legend()
plt.show()
```

```python
# @title Splitting the data into training and test sets and model selection
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, random_state=0)
model_regressor = LinearRegression()
model_regressor.fit(X_train, Y_train)
```

```python
# @title Testing the model
plt.scatter(X, Y, color='purple', label='Data points')
plt.plot(X, model_regressor.predict(X), color='green')
plt.title('Regression Line and Test Data')
plt.legend()
plt.show()
```

```python
# @title Making Predictions
print(X_test) # Testing data - In Hours
Y_pred = model_regressor.predict(X_test) # Predicting the scores
```

```python
# @title Comparing Actual VS Predicted
df = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
df
```

```python
# @title Testing with your own data
hours =9.25
predicted = model_regressor.predict([[hours]])
print("No of Hours Studied",hours)
print("Predicted Score = ", predicted[0])
```

```python
# @title Evaluating the model
print('R-squared Score:', metrics.r2_score(Y_test, Y_pred))
print('Median Absolute Error:', metrics.median_absolute_error(Y_test, Y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))
```
